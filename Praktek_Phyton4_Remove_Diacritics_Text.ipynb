```python
nfdir = pd.read_csv("/content/drive/My Drive/colab/dataset/netflix_directors.csv", encoding='latin-1') #membuka file yang akan dibersihkan, encoding latin-1 fungsinya agar file tersebut dapat dibuka (sebab mengandung teks diacritics yang tidak bisa dibuka oleh phyton)
nfdir.head(15) #untuk menampilkan variabel nfdir 10 baris pertama
```

```python
nfdir['director'] = nfdir['director'].apply(lambda x: unidecode(x)) #script untuk membersihkan teks yang mengandung diacritics pada variabel nfdir pada kolom director
nfdir.head(15) #untuk menampilkan variabel nfdir 15 baris pertama yang telah dibersihkan dari diacritics
```

```python
nfdir.to_csv('/content/drive/My Drive/colab/dataset/netflix_dir.csv', index=False) #export file yang kolomnya telah dibersihkan
```

```python
spesificrow = nfdir[nfdir['director'].str.contains('Mahsun')] #mencari spesifik garis yang mengandung kata "Mahsun" pada kolom director
print(spesificrow)
```

```python
nfdir['director'] = nfdir['director'].str.replace('?', '') #mengganti spesifik karakter '?' menjadi kosong (atau menghapus karakter tsb)
```

```python
nfdir.to_csv('/content/drive/My Drive/colab/dataset/netflix_dir.csv', index=False) #export ulang file yang kolomnya telah dibersihkan
```
