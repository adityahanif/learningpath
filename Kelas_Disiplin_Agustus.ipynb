```python
df['kebersihan'].replace({
    '3. Bersih': "75",
    '4. Sangat Bersih': "100",
    '2. Cukup Bersih': "50",
    '1. Kurang Bersih': "25"
}, inplace=True)
```

```python
df['kerapian'].replace({
    '3. Rapi': "75",
    '4. Sangat Rapi': "100",
    '2. Cukup Rapi': "50",
    '1. Kurang Rapi': "25"
}, inplace=True)
```

```python
df['kelengkapan_fasilitas'] = df['kelengkapan_fasilitas'] * 14.2857
```

```python
df['kelengkapan_seragam'].replace({
    'Semua Lengkap Memakai Seragam & Sepatu': "100",
    'Tidak Lengkap Memakai Seragam & Sepatu': "50"
}, inplace=True)
```

```python
df['keaktifan'].replace({
    '3. Aktif': "75",
    '4. Sangat Aktif': "100",
    '2. Cukup Aktif': "50",
    '1. Kurang Aktif': "25"
}, inplace=True)
```

```python
df['mengingatkan'].replace({
    '1. Membiarkan':"33.333", '3. Terus/Selalu Membangunkan':"99.999", '2. Membangunkan':"66.666"
}, inplace=True)
```

```python
df['menyiapkan_kbm'].replace({
    "0": "50"
}, inplace=True)
```

```python
df['membersihkan_sampah'].replace({
    "0": "50"
}, inplace=True)
```

```python
df['menghapus_papan'].replace({
    "0": "50"
}, inplace=True)
```

```python
df['merapikan_bangku'].replace({
    "0": "50"
}, inplace=True)
```

```python
df['keaktifan_2'].replace({"100":"99.999"
}, inplace=True)
```

```python
df[['kebersihan', 'kerapian', 'kelengkapan_fasilitas', 'kelengkapan_seragam', 'keaktifan', 'keaktifan_2', 'mengingatkan', 'menyiapkan_kbm', 'membersihkan_sampah', 'menghapus_papan', 'merapikan_bangku']] = df[['kebersihan', 'kerapian', 'kelengkapan_fasilitas', 'kelengkapan_seragam', 'keaktifan', 'keaktifan_2', 'mengingatkan', 'menyiapkan_kbm', 'membersihkan_sampah', 'menghapus_papan', 'merapikan_bangku']].astype(float)
```

```python
df.head(5)
```

```python
df.columns
```

```python
df['keaktifan_total'] = (df['keaktifan'] + df['keaktifan_2'] + df['mengingatkan'])/3
df['kepekaan'] = (df['menyiapkan_kbm'] + df['membersihkan_sampah'] + df['menghapus_papan'] + df['merapikan_bangku'])/4
df['kelengkapan'] = (df['kelengkapan_fasilitas'] + df['kelengkapan_seragam'])/2
```

```python
df_new = df[['kelas', 'kebersihan', 'kerapian', 'kelengkapan', 'keaktifan_total', 'kepekaan']]
df_new.head(5)
```

```python
agg = df_new.groupby('kelas').mean()
agg['rata2_total'] = agg.mean(axis=1)
```

```python
agg.sort_values(by='rata2_total', ascending=False)
```

```python
sns.barplot(data=agg['rata2_total'].sort_values(ascending=False), orient='h')
plt.title('Grafik Lomba Kelas Disiplin Agustus [Keseluruhan]')
plt.xlabel('Rata-Rata Akhir')
plt.ylabel('Kelas');
```

```python
sns.barplot(data=agg['kebersihan'].sort_values(ascending=False), orient='h')
plt.title('Grafik Lomba Kelas Disiplin Agustus [Kelas Ter-Bersih]')
plt.xlabel('Kebersihan')
plt.ylabel('Kelas');
```

```python
sns.barplot(data=agg['kerapian'].sort_values(ascending=False), orient='h')
plt.title('Grafik Lomba Kelas Disiplin Agustus [Kelas Ter-Rapi]')
plt.xlabel('Kerapian')
plt.ylabel('Kelas');
```

```python
sns.barplot(data=agg['keaktifan_total'].sort_values(ascending=False), orient='h')
plt.title('Grafik Lomba Kelas Disiplin Agustus [Kelas Ter-Aktif]')
plt.xlabel('Keaktifan')
plt.ylabel('Kelas');
```
