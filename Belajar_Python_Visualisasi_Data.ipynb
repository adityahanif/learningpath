```python
import pandas as pd # this package is for working with large data sets
import matplotlib.pyplot as plt
import numpy as np # this packages contains statistical functions
# Load data
human_data = pd.read_csv('https://content.labxchange.org/labs/datasets/CodeSciLab/HumanHeightWeightData.csv')
human_data
```

```python
human_data.dtypes
```

```python
human_data.describe()
```

```python
print("Tinggi badan minimal: " + str(min(human_data['height_inches'])))
print("Tinggi badan maksimal: " + str(max(human_data['height_inches'])))
print("Berat badan minimal: " + str(min(human_data['weight_pounds'])))
print("Berat badan maksimal: " + str(max(human_data['weight_pounds'])))
print("Rata-rata tinggi badan: " + str(np.mean(human_data['height_inches'])))
print("Rata-rata berat badan: " + str(np.mean(human_data['weight_pounds'])))
print("Median tinggi badan: " + str(np.median(human_data['height_inches'])))
print("Median berat badan: " + str(np.median(human_data['weight_pounds'])))
```

```python
print(human_data.isnull()) #menampilkan cell yang null/kosong
```

```python
fig1 = plt.figure() #membuat grafik
plt.hist(human_data["height_inches"], bins=16) #grafik histogram
plt.title('Distribusi Tinggi Badan')
plt.xlabel('Tinggi Badan (inch)')
plt.ylabel('Jumlah')
fig1.savefig('HeightHistogram1.png') #menyimpan histogram sebagai gambar
#plt.show()

fig2 = plt.figure() #membuat grafik
plt.hist(human_data["weight_pounds"], bins=16) #grafik histogram
plt.title('Distribusi Berat Badan')
plt.xlabel('Berat Badan (pounds)')
plt.ylabel('Jumlah')
fig2.savefig('WeightHistogram2.png') #menyimpan histogram sebagai gambar
#plt.show()
```

```python
fig3 = plt.figure() #membuat grafik
plt.boxplot(human_data["height_inches"]) #grafik histogram
plt.title('Distribusi Tinggi Badan (Garis Merah adalah Median)')
plt.ylabel('Jumlah')
fig3.savefig('HeightBoxplot3.png') #menyimpan histogram sebagai gambar
#plt.show()

fig4 = plt.figure() #membuat grafik
plt.boxplot(human_data["weight_pounds"]) #grafik histogram
plt.title('Distribusi Berat Badan (Garis Merah adalah Median)')
plt.ylabel('Jumlah')
fig4.savefig('WeightBoxplot4.png') #menyimpan histogram sebagai gambar
#plt.show()
```

```python
fig5 = plt.figure()
plt.scatter(human_data["height_inches"], human_data["weight_pounds"])
plt.title("Tinggi Badan vs Berat Badan")
plt.xlabel("Berat Badan (pounds)")
plt.ylabel("Tinggi Badan (inch)")
fig5.savefig("scatterplot.png")
```
